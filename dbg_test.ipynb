```python
# 3E > 0A \n
#去除开头和换行
import re
import os
file1 = open('sequence.fasta','r+')
new_file = 'sequencegenerate.fasta'

if not os.path.exists(new_file):
    os.system(r"touch {}".format(new_file))
file2 = open('sequencegenerate.fasta','r+')
istxt = re.compile(r'>',re.I)
for line in file1.readlines():
    line = line.strip()
    ifstr = re.findall(istxt,line)
    if ifstr:
        line = []
    else:
        file2.write(line)

file1.close()
file2.close()
```
